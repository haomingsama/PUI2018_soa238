```python
import pylab as pl
import pandas as pd
import numpy as np
#imports downloader
from getCitiBikeCSV import getCitiBikeCSV

%pylab inline

import os
#this makes my plots pretty! but it is totally not mandatory to do it
import json
#s = json.load( open(os.getenv ('PUI2016')+"/fbb_matplotlibrc.json") )
#pl.rcParams.update(s)

import scipy as sp
```

```python
datestring = '201501'
getCitiBikeCSV(datestring)
```

```python
df = pd.read_csv(os.getenv("PUIDATA") + "/" + datestring + '-citibike-tripdata.csv')
df.head()
```

```python
# df is the dataframe where the content of the csv file is stored
df['date'] = pd.to_datetime(df['starttime'])
# note that with dataframes I can refer to variables as dictionary keys, 
# i.e. df['starttime'] or as attributes: df.starttime. 
df.head()
```

```python
df.columns
```

```python
df['hour'] = df['date'].dt.hour
df
```

```python
df.hour.unique()
```

```python

day = df.drop(['starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
       'birth year', 'gender', 'date'], axis=1)

day['day'] = df['tripduration'][(df['hour'] > 5) & (df['hour'] <= 18)]
day
```

```python
day = day.dropna()
day
```

```python

night = df.drop(['starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
       'birth year', 'gender', 'date'], axis=1)

night['night'] = night['tripduration'][((night['hour'] >= 0) & (night['hour'] <= 6)) | ((night['hour'] > 18) & (night['hour'] <= 24))]
night
```

```python
night = night.dropna()
night
```

```python
## KS test
```

```python
ks = sp.stats.ks_2samp(day.day, night.night)
print (ks)
```

```python
# The p-value is too large. Which means that there is no correlation between the two groups
```

```python
## Pearson R
```

```python
# taking a sample from our data frame with the same size in order to be able to run the other two tests

day_sample = day.sample(n=1000, frac=None, replace=False, weights=None, random_state=None, axis=None)
night_sample = night.sample(n=1000, frac=None, replace=False, weights=None, random_state=None, axis=None)
```

```python
from scipy.stats import pearsonr

sp.stats.pearsonr(day_sample.day, night_sample.night)
```

```python
## Spearman's test for correlation
```

```python
from scipy.stats import spearmanr

sp.stats.spearmanr(day_sample.day, night_sample.night)
```
